```python
def get_state(state, df2):
    df = df2[df2['State'] == state]
    a = df['Abbreviation']
    if df.size == 0:
        return float('nan')
    else:
        return a.squeeze()
```

```python
def get_state2(state, df2):
    df = df2[df2['Abbreviation'] == state]
    a = df['State']
    if df.size == 0:
        return float('nan')
    else:
        return a.squeeze()
```

```python
def get_pop(state, df2, b):
    state1=get_state(state, state_abv)
    #print(state1)
    df = df2[df2['state'] == state1] 
    a = df[b]
    if df.size == 0:
        return float('nan')
    else:
        return a.squeeze()
```

```python
def primaryCD(state):
    doses_series_new2 = pd.DataFrame(columns = ['date', '50_64','65+'])
    doses_series_new = pd.DataFrame(columns = ['date', '50_64','65+'])
    
    selected=data2[data2['Geography']==state]
    selected_50to64=selected[selected['Age']=='50-64 Years']
    selected_65more=selected[selected['Age']=='65+ Years']
    
    for i in range(1, len(selected_50to64)):
        diff_50to64CD=float((selected_50to64.iloc[i,5]-selected_50to64.iloc[i-1,5])/700)*get_pop(state, data_dem2, '50_64')*(1/3)*0
        diff_65CD=float((selected_65more.iloc[i,5]-selected_65more.iloc[i-1,5])/700)*get_pop(state, data_dem2, '65+')*0
        for j in range(1,7):
            date=date_series_str2[(7*(i-1)+j-1)]
            doses_series_new2 = doses_series_new2.append({'date' : date, '50_64':max(0,diff_50to64CD), '65+':max(0,diff_65CD)}, 
                    ignore_index = True)
        
        doses_series_new = pd.concat([doses_series_new60, doses_series_new2], ignore_index=True)

    
    return doses_series_new
```

```python

vacc_R4_US_CD = primaryCD('United States')
y6 = "vac_US_RSVhigh.csv"
vacc_R4_US_CD.to_csv(y6)

```

```python
import csv

input_csv_file = 'vac_US_RSVhigh.csv'   # Replace with the path to your input CSV file
output_csv_file = 'vac_US_6501high.csv' # Replace with the desired output CSV file path

rows_to_extract = 533

with open(input_csv_file, 'r', newline='') as infile, open(output_csv_file, 'w', newline='') as outfile:
    reader = csv.reader(infile)
    writer = csv.writer(outfile)
    
    for index, row in enumerate(reader):
        if index < rows_to_extract:
            writer.writerow(row)
```
