```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
import math
def my_fun(x, x1, x2):
    return 1/(1 + np.exp(-x1*(x-x2)))
```

```python
import math
def my_fun2(x, a, b):
    return -(a*x+b*x*x)
```

```python
import math
def my_fun3(x, a, b, c):
    return a*np.sin((2*np.pi*x)/238+b)+c
```

```python
BA5 = pd.read_excel('EG5.xlsx')
```

```python
import matplotlib.pyplot as plt

from scipy.optimize import curve_fit
```

```python
ydata = BA5['prev'].to_numpy()
xdata = BA5['days'].to_numpy()
```

```python
plt.plot(xdata, ydata)
```

```python
popt, pcov = curve_fit(my_fun, xdata, ydata, p0 = [0.1, 100], bounds = ([0, 1], [400000, 50000]))
```

```python
X = np.linspace(0, 400, 400)
Y = my_fun(X, popt[0], popt[1])
```

```python
for i in range(100):
    Y[i+100]=Y[i+100]*((250)/((i+250)))
```

```python
plt.plot(X,Y)
plt.plot(xdata,ydata, '*')
```

```python
np.savetxt('EG5.txt', Y)
```

```python
drange1 = pd.date_range("2023-05-13", periods=200, freq="1D")

fig = plt.figure(num=None, figsize=(22, 14), dpi=80, facecolor='w', edgecolor='k')
plt.rcParams.update({'font.size': 40})

plt.plot(drange1, Y[0:200], '--', lw = 8, color='xkcd:lightish blue', label = 'Estimated EG.5 Proportion')
#plt.plot(his_var['date'], his_var['ba4'], '--', lw = 8, color='xkcd:lightish green', label = 'BA.4')
plt.plot(BA5['date'], BA5['prev'], '*', color='xkcd:lightish red', markersize=45, label = 'Reported EG.5 Proportion')
#plt.plot(his_obs['date'], his_obs['ba4'], '*', color='xkcd:lightish green' ,markersize=45)

plt.ylabel('Variant prevalence', fontsize = 40)
#plt.title('BA4 BA5 Competition')
plt.legend(fontsize = 30, loc="upper left")

#drange1[::7], 

plt.xticks( rotation=90)  
plt.ylim([0, 0.99])
plt.grid()
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

```
